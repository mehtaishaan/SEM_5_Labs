```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
#subtask1 - load dataset
capture_v2 = cv2.VideoCapture('VID_2.mp4')
```

```python
#subtask2 - extract frames
fps=10
seconds= 0 
count=1
capture_v2.set(cv2.CAP_PROP_POS_MSEC,seconds*1000)
Framepresent,image = capture_v2.read()
if Framepresent:
    cv2.imwrite("image"+str(count)+".jpg", image)
while Framepresent:
    count+=1
    seconds = seconds + (1/fps)
    seconds = round(seconds, 2)
    capture_v2.set(cv2.CAP_PROP_POS_MSEC,seconds*1000)
    Framepresent,image = capture_v2.read()
    if Framepresent:
        cv2.imwrite("image"+str(count)+".jpg", image)
```

```python
#subtask6 - perform sanity check
def sanity_check(path):
    capture=cv2.VideoCapture(path)
    w=capture.get(cv2.CAP_PROP_FRAME_WIDTH)
    h=capture.get(cv2.CAP_PROP_FRAME_HEIGHT)
    present=1
    Sane=True
    while present:
        present,image=capture.read()
        if(capture.get(cv2.CAP_PROP_FRAME_WIDTH)!=w or capture.get(cv2.CAP_PROP_FRAME_HEIGHT)!=h):
            Sane=False
            break
    if Sane:
        print(path,'Sanity Check PASSED')
    else:
        print(path,'Sanity Check FAILED')
    
        
        
sanity_check('VID_2.mp4')
sanity_check('VID_6.mp4')
sanity_check('VID_14.mp4')
```
