```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv("multiple_linear_data.csv")
data[:5]
```

```python
#fixing zeroes
g2_mean = data['G2'].describe()[1]
g3_mean = data['G3'].describe()[1]
data['G2'] = data['G2'].replace([0],g2_mean)
data['G3'] = data['G3'].replace([0],g3_mean)
```

```python
orig_cols = data.columns
orig_cols
```

```python
#subtask2- Handle categorical values
from sklearn.preprocessing import LabelEncoder

cat = []
enc = LabelEncoder()
for col in orig_cols:
    if data[col].dtype != np.dtype('int64') and data[col].dtype != np.dtype('float64'):
        data[col] = enc.fit_transform(data[col])
r_dum = pd.get_dummies(data['reason'])
data = pd.concat([data, r_dum], axis=1)
data.drop('reason', axis=1, inplace=True)
data
```

```python
d_cp = pd.DataFrame(data)
d_cp.drop('G3', axis=1, inplace=True)
d_cp
```

```python
#subtask3 - Define your X and y
X = d_cp.values
y = np.array(data['G3'])
```

```python
#subtask4 
# Initialize the model
model = LinearRegression()
# Fit the model. Wait! We will complete this step for you ;)
model.fit(X, y)
```

```python
#subtask5 - Print mean squared error
a = model.predict(X)
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y, a)
print(f"\nMSE: {mse}")

if len([i for i, v in enumerate(a) if v < 0]) != 0:
    print('Error in data set. Resolve 0 vals')
```

```python
#subtask6 - Predict on your own data
new_data = np.array([40,  1,  1,  0,  0,  0,  0,  0,  1,  1,  0,  0,  9,  2,  3,  6, 24,
       21,  1,  0,  0,  0]).reshape(1,-1)
print("Predicted grade:",model.predict(new_data))
```
