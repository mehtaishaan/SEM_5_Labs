```python
import numpy as np
```

```python
iter = 5
matrixSize = 3**iter
```

```python
matr = np.zeros((matrixSize, matrixSize),dtype=np.int8)
```

```python
def pehla():
    matr[0][1] = 1
    matr[0][2] = 1
    matr[1][0] = 1
    matr[2][0] = 1
```

```python
n = 1
pehla()
n += 1
for n_iter in range(n,iter):
    step = 3**(n_iter-1)
    for x in range(step):
        for y in range(step):
            matr[x+step][y+step] = matr[x][y]
            matr[x+2*step][y+2*step] = matr[x][y]
    for i in range(step, 2*step):
        matr[0][i] = matr[step][i]
        matr[0][i+step] = matr[step][i]
        matr[i][0] = matr[i][step]
        matr[i+step][0] = matr[i][step]
```

```python
edges =[]
nodes=[]
```

```python
for i in range(matrixSize):
    for j in range(matrixSize):
        if (matr[i][j]==1):
            hello = [int(i),int(j)]
            edges.append(hello)
print(edges)
```

```python
for i in range(1,matrixSize):
    yellow = [int(i),int(i)]
    nodes.append(yellow)
print(nodes)
```

```python
print("nodes are: \n", nodes)
print("edges are: \n", edges)
```

```python
for i in range(len(nodes)):
    print("Node: ",nodes[i])
```

```python
for i in range(len(edges)):
    print("Edge: ",edges[i])
```
