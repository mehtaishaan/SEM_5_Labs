```python
import numpy as np
```

```python
iter = 8
matrixSize = 3**iter
```

```python
m = []
```

```python
apl=np.zeros((10,1))
dia=np.zeros((10,1))
```

```python
for j in range(10):
    matr=np.zeros((matrixSize, matrixSize),dtype=np.bool)
    matr[0][1] = 1
    matr[0][2] = 1
    matr[1][0] = 1
    matr[2][0] = 1
    p=(j/10)+0.05
    for n_iter in range(2,iter):
        step = 3**(n_iter-1)
        for x in range(step):
            for y in range(step):
                    matr[x+step][y+step] = matr[x][y]
                    matr[x+2*step][y+2*step] = matr[x][y]
        for i in range(step, 2*step):
            if np.random.rand()<=p:
                matr[0][i] = matr[step][i]
                matr[i][0] = matr[i][step]
            if np.random.rand()<=p:
                matr[0][i+step] = matr[step][i]
                matr[i+step][0] = matr[i][step]
    m.append(matr)
```

```python
matr=np.array(m)
```

```python
import networkx as nx
```

```python
for i in range (len(matr)):
    a=nx.Graph(matr[i])
    l=list(nx.connected_components(a))
    s=0
    for j in l:
        # print(i)
        s= max(s, nx.diameter(a.subgraph(j)))
    dia[i,0]=s
```

```python
for i in range (len(matr)):
    a=nx.Graph(matr[i])
    l=list(nx.all_pairs_shortest_path_length(a))
    s=0
    for j in l:
        print(i,j[0])
        s=s+sum(j[1].values())
    apl[i,0]=s
    apl
```
