```python
#imports
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

### Using python libraries

```python
#reading and saving grayscale image
img = Image.open('./rgb_sample.jpg').convert('L')
img.save('./out/rgb_sample_grayscale.jpg')
```

```python
img_gray = mpimg.imread('./out/rgb_sample_grayscale.jpg')
img1=mpimg.imread('./rgb_sample.jpg')
plt.figure(1)
plt.subplot(121)
plt.imshow(img1)
plt.subplot(122)
plt.imshow(img_gray , cmap='gray')
plt.show()
```

## Using numpy and matplotlib

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
plt.imshow(rgb2gray(img1), cmap='gray')
```

## Image subtraction/addition etc

```python
i_1 = Image.open('./math_1.jpg')
i_2 = Image.open('./math_2.jpg')
#making size equal
i_2 = i_1.resize((720,480))

i_1 = np.array(i_1)
i_2 = np.array(i_2)

#addition
i_a = i_1 + i_2
#subtraction
i_s = i_1 - i_2
#multiplication
i_m = i_1*i_2

plt.figure(1)
plt.subplot(311)
plt.imshow(i_a)
plt.subplot(312)
plt.imshow(i_s)
plt.subplot(313)
plt.imshow(i_m)


plt.imsave('./out/math_a.jpg', i_a)
plt.imsave('./out/math_s.jpg', i_s)
plt.imsave('./out/math_m.jpg', i_m)
```

## rgb to hsv

```python
from skimage.color import rgb2hsv
hsv_img = rgb2hsv(img1)
plt.imsave('./out/hsv_sample.png', hsv_img)
plt.imshow(hsv_img)
```
